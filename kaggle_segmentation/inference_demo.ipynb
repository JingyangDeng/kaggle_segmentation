```python
from mmseg.apis import init_segmentor, inference_segmentor, show_result_pyplot
from mmseg.core.evaluation import get_palette
import cv2
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np
import os
```

```python
# useful tools
def load_img(path):
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
    img = img.astype('float32') # original is uint16
    img = (img - img.min())/(img.max() - img.min())*255.0 # scale image to [0, 255]
    img = img.astype('uint8')
    return img

def show_img(img, mask=None):
    # clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
   #  img = clahe.apply(img)
#     plt.figure(figsize=(10,10))
    plt.imshow(img, cmap='bone')
    
    if mask is not None:
        # plt.imshow(np.ma.masked_where(mask!=1, mask), alpha=0.5, cmap='autumn')
        plt.imshow(mask, alpha=0.5)
        handles = [Rectangle((0,0),1,1, color=_c) for _c in [(0.667,0.0,0.0), (0.0,0.667,0.0), (0.0,0.0,0.667)]]
        labels = [ "Large Bowel", "Small Bowel", "Stomach"]
        plt.legend(handles,labels)
    plt.axis('off')
```

```python
config_file = '../cache/upernet_originsize_convnext_base_fp16_256x256_160k_kaggle_25d_multilabel_carno_rotate/upernet_originsize_convnext_base_fp16_256x256_160k_kaggle_25d_multilabel_carno_rotate.py'
checkpoint_file = '../cache/upernet_originsize_convnext_base_fp16_256x256_160k_kaggle_25d_multilabel_carno_rotate/best_mDice_iter_160000.pth'
```

```python
# build the model from a config file and a checkpoint file
model = init_segmentor(config_file, checkpoint_file, device='cuda:0')
```

```python
# test a single image
# img_path = '../input/uw-madison-gi-tract-image-segmentation/train/case123/case123_day20/scans/slice_0099_266_266_1.50_1.50.png'
img_path = '../data/mmseg_train_25d_carno/images_25d/case81_day28_89.png'
gt_file_path = '../data/mmseg_train_25d_carno/labels/case81_day28_89.png'
result = inference_segmentor(model, img_path)
```

```python
# show the results
processed_img = load_img(img_path)
res = result[0]
res[res==1]=255

show_img(processed_img, res)
```

```python
# show gt
gt = cv2.imread(gt_file_path, cv2.IMREAD_UNCHANGED)
gt[gt==1]=255
show_img(processed_img, gt)
```
